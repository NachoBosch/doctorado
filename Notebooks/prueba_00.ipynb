* Hay 23351 genes en el conjunto de datos.

* El genotipo de los ratones con enfermedad de Huntington es ployQ 111.

```python
#Leemos el dataset con pandas
import pandas as pd
df = pd.read_csv('../Data/striatum_tissue_HD_mice_rna_seq.csv')
df.head()
```

```python
#Nos aseguramos de que no existan valores nulos en el dataset
df.isnull().sum()
```

```python
#Verificamos que no hayan genes repetidos
df.genename.unique().shape
```

```python
#Observamos los estad√≠sticos de cada muestra de rat√≥n
df.describe().T
```

```python
#Analizamos el ajuste de bondad para cada muestra
from scipy import stats
for i in df.columns[1:]:
    print(f"Shapiro de {i}: {stats.shapiro(df[i])}")
    print(f"Kolmogorov de {i}: {stats.kstest(df[i], stats.norm.cdf)}\n")
```
