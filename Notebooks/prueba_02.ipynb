# GSE 3790

```python
import pandas as pd
```

```python
df = pd.read_json('../Data/GSE3790-version2/aggregated_metadata.json')
grade_hd_value = df['samples']['GSM86788']['refinebio_annotations'][0]['characteristics_ch1'][0]
print(grade_hd_value)
```

```python
df_HD = []
df_samples = df['samples'] 
for keys, values in df_samples.items():
    if keys == 'GSE3790':
        pass
    else:
        hd_grade = df_samples[keys]['refinebio_annotations'][0]['characteristics_ch1'][0]
        hd_grade = hd_grade.split()
        if "control," in hd_grade:
            print(keys,"control")
            df_HD.append({'Samples':keys,'Grade':6}) 
        elif "grade" in hd_grade:
            grade = hd_grade[hd_grade.index('grade') + 1][:-1]
            df_HD.append({'Samples':keys,'Grade':int(grade)})
df_HD = pd.DataFrame(df_HD)
```

```python
df_HD.info()
```

```python
df_HD.to_csv('../Data/df_HD_grade.csv')
```
